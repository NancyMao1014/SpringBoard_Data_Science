### 3. Correlation between independent variables

Here again I assume correlation of two variables is strong if the absolute value of their correlation coefficient is higher than or equal to 0.5

```python
#new corr. matrix without the dependent variable
corr_matrix_indep = corr_matrix.drop(index='pricePerSqMeter', columns='pricePerSqMeter')
```

```python
# find strongly correated independing variable pairs
corr_indep = pd.DataFrame()
for i, col in enumerate(corr_matrix_indep.index):
    for j, idx in enumerate(corr_matrix_indep.index):
        if i > j:
            corr = corr_matrix_indep.loc[idx,col]
            if  (corr >= .5) | (corr <= -.5):
                _, p_val = stats.pearsonr(df[col], df[idx])
                corr_indep.loc[idx+'_vs_'+col, 'corr'] = round(corr,4)
                corr_indep.loc[idx+'_vs_'+col, 'p_val'] = p_val
                
print(corr_indep.sort_values(by=['corr'], ascending=False)) 
```

The number of bedrooms, living rooms, and bathrooms shows a high positive correlation with the square meters of the house. But the number of kitchens does not have a strong positive correlation with the square meters.

### 4. Relationship between house price and categorical independent variables

From the previous EDA analysis, we discovered that among the three binary variables (elevator, fiveYearsProperty and subway), only subway seems have a big impact on house price. Let's test if the differences of price caused by those the binary variable are significant.

I will use three types of tests: Permutation Test, Wmann-whitney test and Welch's t test

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y

def diff_of_means(data_1, data_2):
    """Difference in means of two arrays."""

    # The difference of means of data_1, data_2: diff
    diff = np.abs(np.mean(data_1) - np.mean(data_2))
    newDiff = diff.item()

    return newDiff


def permutation_sample(data1, data2):
    """Generate a permutation sample from two data sets."""

    # Concatenate the data sets: data
    data = np.concatenate((data1,data2))

    # Permute the concatenated array: permuted_data
    permuted_data = np.random.permutation(data)

    # Split the permuted array into two: perm_sample_1, perm_sample_2
    perm_sample_1 = permuted_data[:len(data1)]
    perm_sample_2 = permuted_data[len(data1):]

    return perm_sample_1, perm_sample_2


def draw_perm_reps(data_1, data_2, func, size=1):
    """Generate multiple permutation replicates."""

    # Initialize array of replicates: perm_replicates
    perm_replicates = np.empty(size)

    for i in range(size):
        # Generate permutation sample
        perm_sample_1, perm_sample_2 = permutation_sample(data_1, data_2)

        # Compute the test statistic
        perm_replicates[i] = func(perm_sample_1, perm_sample_2)

    return np.array(perm_replicates)


def bootstrap_replicate_1d(data, func):
    """Generate bootstrap replicate of 1D data"""
    replicatedArray = np.random.choice(data.flatten(), size=len(data))
    return func(replicatedArray)

def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] =  bootstrap_replicate_1d(data,func)  

    return bs_replicates
```

#### Elevator

```python
group1=df[df.elevator == '1.0']['pricePerSqMeter']
group0=df[df.elevator == '0.0']['pricePerSqMeter']

x1, y1 = ecdf(group1)
x0, y0 = ecdf(group0)


# Plot all ECDFs on the same plot
fig1 = plt.plot(x0, y0, marker = '.', linestyle = 'none')
fig1 = plt.plot(x1, y1, marker = '.', linestyle = 'none')
fig1 =plt.legend(('No Elevator', 'With Elevator'),loc='lower right')
fig1 = plt.xlabel('Price Per Sq. Meter')
fig1 = plt.ylabel('ECDF')
fig1 = plt.title('ECDF of Price Per Sq. Meter by house with elevator vs. house without elevator')
plt.show()

# Box Plot
fig2 = sns.boxplot(x='elevator', y='pricePerSqMeter', data=df)
fig2 = plt.title(' Box Plot: Price Per Sq. Meter by house with elevator vs. house without elevator')
plt.show()
```

```python
# Hypothesis Test 1 - Permutation Test

# Simulating the null hypothesis that two quantities(groups) are identically distributed.
# If the distribution of the two variables are identical, they should have the same means. 
# The difference of the means is our test statistic.
# we resample the data by shuffling lables.

# p-value in this case is measuring the proportion of the permutation replicates that are at least as 
# extreme as the the actual difference of means.

# Compute difference of mean: empirical_diff_means
empirical_diff_means = diff_of_means(group0, group1)

print('Empirical Diff of Mean:', empirical_diff_means)

# Draw 10,000 permutation replicates: perm_replicates
perm_replicates= draw_perm_reps(group0, group1, diff_of_means, size=10000)

fig3 = plt.hist(perm_replicates, bins=100)
fig3 = plt.xlabel('Permutated Replciates: Diff of Means')
fig3 = plt.ylabel('Count out of 10,000 Draws')
fig3 = plt.title('Histogram of Permutated Replicates of Diff of Means - Elevator')

plt.show()

# Compute p-value
p_value = len(perm_replicates[perm_replicates >= empirical_diff_means]) / len(perm_replicates)

# Print the result
print('Proportion of permutation replicates with value at least as extreme as the empirical diff of means')
print('p-value = %.4f' % p_value)
```

```python
# Hypothesis Test 2 - Two Sample Bootstrap Hypothesis Test for Differences of Means

# Simulating the null hypothesis that the two quantities(groups) have the same means.
# We shift both arrays of data to have the same mean 
# and generate boostrap replicates which are the diffrences of means


# Compute mean of all forces: mean_concat
group_concat = pd.concat([group0, group1])
group_concat = np.array(group_concat)
mean_concat = np.mean(group_concat)
print('Mean of Concatenated Data:', mean_concat)


# Shift both arrays to have the same mean
group0_shifted = group0 - np.mean(group0) + mean_concat
group1_shifted = group1 - np.mean(group1) + mean_concat

group0_shifted = np.array(group0_shifted)
group1_shifted = np.array(group1_shifted)

print('\n')
print('Mean of shifted group 0',group0_shifted.mean())
print('Mean of shifted group 1',group1_shifted.mean())

fig4 = plt.hist(group0_shifted, alpha=0.5, bins = 100)
fig4 = plt.hist(group1_shifted, alpha=0.5, bins = 100)

fig4 = plt.xlabel('Shifted Array Values of House Pricing')
fig4 = plt.ylabel('Count')
fig4 = plt.legend(('No Elevator', 'With Elevator'),loc='lower right')
fig4 = plt.title('Histogram of Shifted Arrays of House Pricing for Boostrap Hypothesis - Elevator')

plt.show()

# Compute 10,000 bootstrap replicates from shifted arrays
bs_replicates_0 = draw_bs_reps(group0_shifted, np.mean, size=10)
bs_replicates_1 = draw_bs_reps(group1_shifted, np.mean, size=10)

# Get replicates of difference of means: bs_replicates
bs_replicates = bs_replicates_1 - bs_replicates_0

# Compute and print p-value:
p_value = np.sum(bs_replicates >= empirical_diff_means) / len(bs_replicates)

print('\n')
print('Proportion of bootstrap replicates with value at least as extreme as the empirical diff of means')
print('p-value = %.4f' % p_value)

fig5 = plt.hist(bs_replicates, alpha=0.5)
fig5 = plt.xlabel('Bootstrap Replicates')
fig5 = plt.ylabel('Count')
fig5 = plt.title('Histogram of Bootstrap Replicates & Count')

plt.show()
```

```python
#Hypothesis Test 3: Mann-Whitney Test
#A Nonparametric test used to test whether two samples are likely to derive from the same population.
print(stats.mannwhitneyu(group0, group1))


#Hypothesis Test 4: Welch's t-test
#A two-sample location test which is used to test the hypothesis that two populations have equal means 
#when the two samples have unequal variances and/or unequal sample sizes.
print(stats.ttest_ind(group0, group1, equal_var = False))
```

All four tests all show a very small p-value which means that the impact of elevator on house price is statistically significant
